```python
import scipy.signal
import numpy as np
import IPython.display as ipd
import math
import matplotlib.pyplot as plt

plot_library = 'matplotlib'
```

```python

# fs = 44100

# t = np.linspace(0,1,44100)

# r = 0.004
# u = 0.01

def play_bubble(r, xi, u, fs):
    t = np.linspace(0,1,44100)

    ## set to 1 so that its normalized
    a = 1
    #a = r * math.sqrt(r) * u

    d = (0.013 / r) + (0.0072 * math.pow(r, -3/2))

    #xi = 0.1
    sigma = xi * d
    f = 1 + sigma * t
    f = (3/r) * f

    out_arr = a * np.sin(2 * np.pi * f * t) * (np.power(math.e, -d * t)).astype(np.float32)

    plt.plot(out_arr[0:5000])

    return out_arr

fs = 44100
## Each of the bubble noises from the study. Xi has not been added since its can be changed with the second variable.
out_arr1 = play_bubble(0.01, 0.05, 0.01, fs)
out_arr2 = play_bubble(0.007, 0.05, 0.01, fs)
out_arr3 = play_bubble(0.004, 0.05, 0.01, fs)
out_arr4 = play_bubble(0.002, 0.05, 0.01, fs)
out_arr5 = play_bubble(0.001, 0.05, 0.01, fs)
out_arr6 = play_bubble(0.0005, 0.05, 0.01, fs)
out_arr7 = play_bubble(0.0003, 0.05, 0.01, fs)
ipd.Audio(out_arr1, rate=fs, normalize=False)
```

```python
def setRate(total_rate, bubble_rate):
    rate_sum = sum(bubble_rate)

    return ((total_rate * bubble_rate) / rate_sum)



bubble_rate = np.array([0.2, 0.3, 0.5, 0.6, 0.2])
total_rate = 10

rate = setRate(total_rate, bubble_rate)
plt.plot(rate)
```
